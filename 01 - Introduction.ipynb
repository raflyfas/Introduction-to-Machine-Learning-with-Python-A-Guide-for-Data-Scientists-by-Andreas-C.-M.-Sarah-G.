# NumPy

NumPy adalah salah satu pustaka (library) utama untuk komputasi ilmiah dalam Python. Pustaka ini menyediakan kemampuan untuk mengolah array multidimensi, berbagai fungsi matematika tingkat tinggi (seperti aljabar linier dan transformasi Fourier), serta generator angka acak semu (pseudorandom). Dalam machine learning, data biasanya direpresentasikan sebagai array NumPy, sehingga pustaka ini menjadi fondasi yang sangat penting.

```python
# Membuat array 2 dimensi dengan NumPy
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Bentuk Array (ndarray):\n", arr)
```

# SciPy

SciPy merupakan kumpulan fungsi dan algoritma untuk komputasi ilmiah di Python. Library ini menyediakan modul untuk aljabar linier tingkat lanjut, optimasi fungsi matematika, pemrosesan sinyal, fungsi matematika khusus, serta distribusi statistik. Banyak algoritma di scikit-learn yang dibangun menggunakan fungsi-fungsi dari SciPy. Salah satu modul SciPy yang sangat krusial bagi kita adalah `scipy.sparse`. Modul ini menyediakan matriks jarang (sparse matrices), yang digunakan untuk merepresentasikan array 2 dimensi jika sebagian besar elemennya bernilai nol secara efisien, sehingga menghemat penggunaan memori.

```python
from scipy import sparse

# Membuat matriks identitas 2D menggunakan NumPy
matriks_identitas = np.eye(4)
print("Matriks NumPy 2D:\n", matriks_identitas)

# Mengubah matriks tersebut menjadi format Sparse Matrix (CSR) di SciPy
# Format ini hanya menyimpan entri yang bernilai bukan nol (nonzero)
matriks_jarang = sparse.csr_matrix(matriks_identitas)
print("\nMatriks Sparse CSR SciPy:\n", matriks_jarang)
```

```python
# Membuat representasi COO (Coordinate Format) secara langsung
data_elemen = np.ones(4)
indeks_baris = np.arange(4)
indeks_kolom = np.arange(4)

matriks_coo = sparse.coo_matrix((data_elemen, (indeks_baris, indeks_kolom)))
print("Matriks Sparse dalam representasi COO:\n", matriks_coo)
```

# Mengenal Dataset 


Untuk studi kasus pertama ini, kita akan menggunakan dataset bunga Iris, salah satu dataset paling klasik dan populer dalam statistik dan pembelajaran mesin (machine learning). Dataset ini sudah terintegrasi langsung dalam modul `datasets` pada scikit-learn. Kita dapat memuatnya secara mudah dengan memanggil fungsi `load_iris`:

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# Eksplorasi atribut dasar dataset
print("Atribut kunci (keys) dataset:", list(iris_dataset.keys()))
print("Nama target (spesies):", iris_dataset['target_names'])
print("Daftar fitur pengukuran:", iris_dataset['feature_names'])
```

```python
# Memeriksa tipe data dan dimensi array fitur dan target
print("Tipe data fitur (data):", type(iris_dataset['data']))
print("Dimensi array fitur (data):", iris_dataset['data'].shape)
print("Tipe data target (target):", type(iris_dataset['target']))
print("Dimensi array target (target):", iris_dataset['target'].shape)
```

```python
# Deskripsi singkat dataset
print("Potongan Deskripsi Dataset:\n", iris_dataset['DESCR'][:193], "\n...")

# Menampilkan 5 sampel pertama
print("\n5 baris pertama data sampel:\n", iris_dataset['data'][:5])

# Menampilkan label seluruh sampel
print("\nLabel kelas target:\n", iris_dataset['target'])
```

Dapat dilihat bahwa array tersebut menampung data pengukuran dari 150 bunga yang berbeda. Dalam istilah machine learning, setiap objek atau baris data disebut sebagai sampel (samples), sedangkan kolom atau properti pengukurannya disebut sebagai fitur (features). Dimensi (shape) dari array data ini adalah jumlah sampel dikalikan jumlah fitur. Format/konvensi dimensi data seperti ini merupakan standar yang wajib diikuti saat menggunakan library scikit-learn.

Berdasarkan data 5 baris pertama tersebut, kita bisa melihat bahwa kelima sampel memiliki lebar mahkota bunga (petal width) sebesar 0.2 cm, dan bunga pertama memiliki kelopak bunga (sepal length) terpanjang sebesar 5.1 cm.

# Memisahkan Data Latih (Training) dan Data Uji (Testing)

Tujuan akhir kita adalah membuat model machine learning yang bisa memprediksi spesies bunga Iris berdasarkan data pengukuran baru. Sebelum menggunakannya untuk data baru, kita harus mengukur performanya terlebih dahulu untuk mengetahui seberapa akurat prediksi model tersebut.

Sangat penting untuk diingat bahwa kita tidak boleh menggunakan data latih (data yang dipakai membuat model) untuk mengevaluasi kinerja model itu sendiri. Jika kita melakukannya, model bisa dengan mudah "menghafal" seluruh data latih, sehingga prediksinya akan selalu terlihat sempurna pada data tersebut. Namun, proses "menghafal" ini tidak memberi tahu kita apakah model dapat melakukan generalisasi dengan baik jika dihadapkan pada data baru yang belum pernah dilihat sebelumnya.

---

Oleh karena itu, cara terbaik untuk mengevaluasi performa model adalah dengan mengujinya menggunakan data baru yang sudah memiliki label. Biasanya, dataset awal kita (150 sampel bunga Iris) akan dibagi menjadi dua bagian:
1. **Data Latih (Training Set)**: Bagian data yang digunakan khusus untuk melatih dan membangun model machine learning.
2. **Data Uji (Test Set / Hold-out Set)**: Bagian data sisa yang digunakan untuk menguji performa prediksi model setelah proses latihan selesai.

```python
from sklearn.model_selection import train_test_split

# Membagi data dengan proporsi 75% train dan 25% test
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0
)

# Menampilkan dimensi data hasil split
print(f"Dimensi X_train (fitur latih): {X_train.shape}")
print(f"Dimensi y_train (label latih): {y_train.shape}")
print(f"Dimensi X_test (fitur uji)   : {X_test.shape}")
print(f"Dimensi y_test (label uji)   : {y_test.shape}")
```

Di scikit-learn, proses pengacakan (shuffling) dataset dan pembagian data ini dapat dilakukan secara otomatis menggunakan fungsi `train_test_split`. Secara default, fungsi ini akan memisahkan 75% data beserta labelnya sebagai data latih, dan sisa 25% data beserta labelnya dijadikan sebagai data uji. Meskipun proporsi pembagian data latih dan uji ini relatif bebas ditentukan, pembagian 75:25 atau 80:20 merupakan standar (rule of thumb) yang umum digunakan.

Dalam pemrograman scikit-learn, data fitur biasanya disimbolkan dengan huruf kapital $X$, sedangkan label kelas menggunakan huruf kecil $y$. Simbol ini merujuk pada rumus matematika standar $f(x) = y$, di mana $x$ adalah input dan $y$ adalah outputnya. Penggunaan huruf kapital $X$ dikarenakan data fitur merupakan array dua dimensi (matriks), sedangkan huruf kecil $y$ mewakili target yang berbentuk array satu dimensi (vektor).

```python
# Membuat DataFrame dari X_train dengan nama kolom dari fitur Iris
df_iris = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# Visualisasi matriks diagram tebar (scatter matrix) dengan pewarnaan berdasarkan kelas target y_train
pd.plotting.scatter_matrix(
    df_iris, c=y_train, figsize=(15, 15), marker='o',
    hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3
)
plt.suptitle("Matriks Scatter Plot Data Latih Iris", y=0.9, fontsize=16)
plt.show()
```

Melalui grafik sebaran (scatter plot) tersebut, kita dapat melihat bahwa ketiga spesies bunga terpisah secara berkelompok dengan cukup jelas berdasarkan pengukuran sepal dan petal. Hal ini menunjukkan indikasi awal yang baik bahwa model machine learning akan mampu mempelajari pola tersebut dengan mudah untuk membedakan ketiganya.

# K-Nearest Neighbors (KNN)


```python
from sklearn.neighbors import KNeighborsClassifier

# Inisialisasi model KNN dengan parameter tetangga terdekat K = 1
model_knn = KNeighborsClassifier(n_neighbors=1)

# Melatih model menggunakan data latih (X_train dan y_train)
model_knn.fit(X_train, y_train)
print("Model KNN berhasil dibangun dan dilatih:\n", model_knn)
```

```python
# Definisikan sampel baru bunga Iris untuk diprediksi
sampel_baru = np.array([[5, 2.9, 1, 0.2]])
print("Sampel baru:", sampel_baru)
print("Bentuk sampel baru:", sampel_baru.shape)

# Melakukan prediksi menggunakan model KNN yang sudah dilatih
prediksi_kelas = model_knn.predict(sampel_baru)
nama_spesies = iris_dataset['target_names'][prediksi_kelas]

print("\nHasil Prediksi Kelas:", prediksi_kelas)
print("Nama Spesies Hasil Prediksi:", nama_spesies[0])
```

```python
# Melakukan prediksi untuk seluruh data uji (X_test)
y_prediksi = model_knn.predict(X_test)
print("Hasil prediksi untuk data uji:\n", y_prediksi)

# Evaluasi akurasi menggunakan metode rata-rata kecocokan manual
akurasi_manual = np.mean(y_prediksi == y_test)
# Evaluasi akurasi menggunakan method score() dari model KNN
akurasi_model = model_knn.score(X_test, y_test)

print(f"\nAkurasi Evaluasi Uji (Metode Manual)  : {akurasi_manual:.2f}")
print(f"Akurasi Evaluasi Uji (Metode KNN.score) : {akurasi_model:.2f}")
```

```python
# Rangkuman lengkap proses dari split data, fitting model, hingga evaluasi akurasi
X_train_summary, X_test_summary, y_train_summary, y_test_summary = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0
)

knn_summary = KNeighborsClassifier(n_neighbors=1)
knn_summary.fit(X_train_summary, y_train_summary)

skor_akhir = knn_summary.score(X_test_summary, y_test_summary)
print(f"Akurasi akhir model KNN terintegrasi: {skor_akhir:.2f}")
```
