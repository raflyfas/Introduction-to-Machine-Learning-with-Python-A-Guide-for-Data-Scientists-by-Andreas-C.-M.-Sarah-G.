## Business Metric dan Machine Learning Metric

Machine learning metric seperti accuracy, precision, recall, atau R² hanya mengukur performa model secara teknis. Namun, dalam dunia nyata, yang lebih penting adalah dampak akhirnya.

Misalnya, pada sistem fraud detection, metrik teknis dapat berupa recall untuk mendeteksi transaksi fraud. Akan tetapi, metrik bisnisnya bisa berupa penurunan kerugian finansial. Model yang secara teknis terlihat bagus belum tentu memberikan keuntungan jika biaya implementasinya lebih besar daripada manfaatnya.

## Machine Learning sebagai Feedback Loop

Modeling bukan proses satu arah. Dalam praktiknya, hasil model sering memberi informasi baru tentang data. Kesalahan model dapat menunjukkan bahwa data kurang lengkap, label tidak konsisten, fitur belum cukup representatif, atau masalah perlu dirumuskan ulang.

Siklus umum proyek machine learning:

1. Menentukan masalah.
2. Mengumpulkan data.
3. Membersihkan dan memahami data.
4. Membuat model awal.
5. Mengevaluasi kesalahan model.
6. Memperbaiki data, fitur, atau definisi masalah.
7. Mengulang proses hingga solusi cukup baik.

## Humans in the Loop

Tidak semua keputusan harus sepenuhnya otomatis. Dalam beberapa kasus, sistem machine learning dapat menangani kasus yang mudah, sementara kasus yang sulit atau berisiko tinggi tetap diberikan kepada manusia.

Contohnya:
- Pada aplikasi medis, dokter tetap memegang keputusan akhir.
- Pada sistem moderasi konten, model dapat menandai konten mencurigakan untuk diperiksa moderator.
- Pada fraud detection, transaksi dengan risiko tinggi dapat dikirim ke tim manual review.

Pendekatan ini disebut **humans in the loop**.

Humans in the loop dapat mengurangi risiko dan meningkatkan kepercayaan terhadap sistem.

## Dari Prototype ke Production

Perbedaan umum prototype dan production:

## Testing Production Systems

Evaluasi pada test set disebut offline evaluation. Ini penting, tetapi belum cukup untuk sistem yang digunakan langsung oleh user. Ketika model masuk ke production, perilaku user dapat berubah karena model memengaruhi pengalaman mereka.

Untuk sistem user-facing, pengujian online seperti A/B testing sering digunakan.

## A/B Testing

A/B testing membandingkan dua versi sistem pada kelompok user yang berbeda. Sebagian user melihat sistem A, sebagian lainnya melihat sistem B. Setelah periode tertentu, metrik keberhasilan dibandingkan.

Contoh:
- Sistem A menggunakan rekomendasi lama.
- Sistem B menggunakan model rekomendasi baru.
- Metrik yang dibandingkan bisa click-through rate, conversion rate, atau retention.

A/B testing membantu melihat dampak model di kondisi nyata.

## Contoh Simulasi A/B Testing

Contoh berikut menunjukkan simulasi sederhana perbandingan conversion rate antara dua versi sistem.

```python
ab_test = pd.DataFrame({
    "Group": ["A - Old System", "B - New Model"],
    "Users": [10000, 10000],
    "Conversions": [820, 910]
})

ab_test["Conversion Rate"] = ab_test["Conversions"] / ab_test["Users"]

ab_test
```

## Analisis A/B Testing

Pada simulasi di atas, sistem B memiliki conversion rate lebih tinggi. Namun, dalam praktik nyata, keputusan tidak hanya dilihat dari angka mentah. Perlu dipertimbangkan juga signifikansi statistik, durasi eksperimen, segmentasi user, dan kemungkinan efek samping.

## Membuat Estimator Sendiri

Kadang, preprocessing khusus yang dibutuhkan tidak tersedia langsung di scikit-learn. Jika preprocessing tersebut bergantung pada data dan ingin digunakan dalam pipeline atau cross-validation, maka lebih aman membuat custom estimator.

Scikit-learn menyediakan `BaseEstimator` dan `TransformerMixin` untuk membuat transformer sendiri yang kompatibel dengan Pipeline.

```python
class AddConstantFeature(BaseEstimator, TransformerMixin):
    def __init__(self, value=1.0):
        self.value = value

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        constant_column = np.full((X.shape[0], 1), self.value)
        return np.hstack([X, constant_column])
```

## Menggunakan Custom Transformer dalam Pipeline

Transformer buatan sendiri dapat digunakan seperti transformer scikit-learn lainnya. Contoh berikut menambahkan satu fitur konstan ke dataset sebelum data distandardisasi dan dimasukkan ke Logistic Regression.

```python
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=42,
    stratify=cancer.target
)

pipe = Pipeline([
    ("add_constant", AddConstantFeature(value=1.0)),
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression(max_iter=5000))
])

pipe.fit(X_train, y_train)

print(f"Test score: {pipe.score(X_test, y_test):.3f}")
```

## Cross-Validation dengan Custom Transformer

Karena transformer sudah kompatibel dengan scikit-learn, pipeline juga dapat digunakan pada cross-validation.

```python
scores = cross_val_score(pipe, cancer.data, cancer.target, cv=5)

print("Cross-validation scores:", scores)
print(f"Mean CV score: {scores.mean():.3f}")
```
